```python
gender=pd.get_dummies(df['Gender'], dtype=int)
pass_division=pd.get_dummies(df['Pass_Division'], dtype=int)
```

```python
df2=df.drop(['Gender', 'Pass_Division'], axis=1)
```

```python
df3=df2.join([gender, pass_division])
df3
```

```python
month=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df3['Birth_Month']=pd.Categorical(df3['Birth_Month'], month, ordered=True)
```
